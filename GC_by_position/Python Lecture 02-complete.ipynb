```python
import gzip
fin = gzip.open('test_data/heli.R1s.fq.gz', 'rb') #rb means binary mode

def avgread(seqfile):
    seq_lengths = []
    for count, line in enumerate(seqfile):
        line = line.strip()
        if count % 4 == 1:
            seq_lengths.append(len(line))
    average = sum(seq_lengths)/ float(len(seq_lengths))
    return average

avg = avgread(fin)
print avg
    
```

```python
import gzip
fin = gzip.open('test_data/heli.R1s.fq.gz', 'rb')

def readlength(seqfile):
    sum = 0
    numlines = 0
    for line in enumerate(seqfile):
        count, txt = line
        txt = txt.strip()
        if count % 4 == 1:
            sum += len(txt)
            numlines += 1
    average = float(sum) / numlines
    return average

readlength(fin)
```

### Calculate the GC content for each sequence

* Think about how you can use a dictionary to store the counts

```python
import gzip
fin = gzip.open('test_data/heli.R1s.fq.gz', 'rb')

def gccount(seqfile):
    gc = {}
    for line in enumerate(seqfile):
        count, txt = line
        if count % 4 == 1:
           gc[count] = 0
           for i in range(len(txt)):
               if txt[i] == 'G' or txt[i] == 'C' or txt[i] == 'g' or txt[i] == 'c':
                   gc[count] += 1
    return gc

gcc = gccount(fin)

gcc.values()
```

```python
import numpy
numpy.average(gcc.values())
```

## calculate GC content for each base position across all reads in a fastq file

```python
import gzip
fin = gzip.open('test_data/heli.R1s.fq.gz', 'rb')

#We know there are 10,000 sequences of average length 90
seqlength = 90
numseqs = 10000.


def gccount_by_base(seqfile):
    GC_by_pos = [0]*seqlength
    for line in enumerate(fin):
        count, txt = line
        txt = txt.strip()
        if count % 4 == 1:
            for i in range(len(txt)):
                if txt[i] == 'G' or txt[i] == 'C' or txt[i] == 'g' or txt[i] == 'c':
                    GC_by_pos[i] += 1
    for j in range(len(GC_by_pos)):
        GC_by_pos[j] = GC_by_pos[j]/numseqs
    return GC_by_pos

GC = gccount_by_base(fin)
print GC
```
